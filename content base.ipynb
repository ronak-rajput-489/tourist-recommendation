## word steming.....

```python
#!pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(x):
    l=[]
    for i in x.split():
        l.append(ps.stem(i))
    return ' '.join(l)
```

```python
def recomd(famous):
    indexs=df_2[df_2['famousefor']==famous].index[0]
    dist=similarity[indexs]
    place_list=sorted(list(enumerate(dist)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in place_list:
        print(df_2.iloc[i[0]].famousefor)
```
