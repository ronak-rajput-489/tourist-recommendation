```python
def recomd_city(city):
    indexes=np.where(df_final.index==city)[0][0]
    similar_city=sorted(list(enumerate(similarity_score[indexes])),key=lambda x:x[1],reverse=True)[1:6]
    
    for i in similar_city:
        print(df_final.index[i[0]])
```
