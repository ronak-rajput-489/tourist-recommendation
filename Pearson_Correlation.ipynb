```python
df_final = df
```

```python
df_final = df_final.fillna(0)
df_final
```

### Correlation Recommedation System

```python
df_final.columns
```

```python
vishakhapatnam_rating = df_final["Visakhapatnam"]
similar_to_vishakhapatnam = df_final.corrwith(vishakhapatnam_rating)
corr_vish = pd.DataFrame(similar_to_vishakhapatnam, columns = ['Pearson correlation'])
recommend = corr_vish.sort_values(ascending = False, by  = 'Pearson correlation')[1:7]
recommend
```

```python
def recommend(city):
    city_rating = df_final[city]
    similar_to_city = df_final.corrwith(city_rating)
    corr_city = pd.DataFrame(similar_to_city, columns = ['Correlation'])
    recommend = corr_city.sort_values(ascending = False, by = ['Correlation'])[1:7]
    return recommend
```

```python
recommend('Visakhapatnam')
```
